```python
import pandas as pd
import os
import json
import requests
from pprint import pprint
```

```python
file = os.path.join('airports_all_2015.csv')
airports_df = pd.read_csv(file)
airports_df.head()
```

```python
def lat_long(iata):

    url = "https://cometari-airportsfinder-v1.p.rapidapi.com/api/airports/by-code"

    querystring = {"code": f"{iata}"}

    headers = {
        'x-rapidapi-host': "cometari-airportsfinder-v1.p.rapidapi.com",
        'x-rapidapi-key': "8cc177b195msh4e4f4ff4e0cae0fp10ed89jsn60e0207e5c4d"
        }

    response = requests.get(url, headers=headers, params=querystring).json()

    latitude = response[0]['location']['latitude']
    longitude = response[0]['location']['longitude']

    return latitude, longitude
```

```python
# Find Latitude and Longitude of each airport
lat_list = []
long_list = []

for i in range(len(airports_df)):
    iata = airports_df['Code (IATA/ICAO)'][i]
    latitude, longitude = lat_long(iata)
    lat_list.append(latitude)
    long_list.append(longitude)
```

```python
# Add lat, long to dataframe
airports_df['Lat'] = lat_list
airports_df['Long'] = long_list
airports_df.dropna(inplace=True)
# airports.head()
```

```python
# Check if each airport is international
file = os.path.join('airports_intl_2016.csv')
intl_df = pd.read_csv(file)
# intl_df.head()

intl = intl_df['Code (IATA/ICAO)']
intl_bool = []

for index, row in airports_df.iterrows():
    iata = row['Code (IATA/ICAO)']
    if iata in intl.unique():
        intl_bool.append("International")
    else:
        intl_bool.append("Domestic")

airports_df['Type'] = intl_bool
airports_df.head()
```

```python
file = os.path.join('airports.csv')
airports_df.to_csv(file, index=False, header=True)
```
